```python

import numpy as np
from KNN_Classifier import KNN
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt(open('Outlier.txt',"rt"),delimiter=",")
np.random.shuffle(data)      #randomnly shuffles the data
```

```python

#Created instance of class
train=KNN()
```

```python

#Separates data into X and Y and normalizes X
X=data[:,:2]
Y=data[:,-1:].flatten()
X=train.normalize(X)
```

```python
#dividing into training data and test data in the ratio 3:1
test_len=X.shape[0]//4
X_test=X[:test_len,:]
X_train=X[test_len:,:]

Y_test=Y[:test_len]
Y_train=Y[test_len:]
```

```python
# VISUALIZATION of training data
plt.scatter(X_train[:,0],X_train[:,1],c=Y_train)
plt.xlabel("training dataset 1")
plt.ylabel("training dataset 2")
plt.title("Training Data")
plt.show()
```

```python
#fits the training data
train.fit(X_train,Y_train)
```

```python
#predicts the values of Y based on X_test data
Y_pred = train.pred(X_test)
```

```python
# VISUALIZATION of predicted data
plt.scatter(X_test[:,0],X_test[:,1],c=Y_pred)
plt.xlabel("Testing dataset 1")
plt.ylabel("Testing dataset 2")
plt.title("Testing Data")
plt.show()#Calculates the accuracy of my model
Acc = train.Accuracy(Y_test,Y_pred)
print("Accuracy of my model : ",Acc,"%")
```
