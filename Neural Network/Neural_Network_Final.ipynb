```python

import numpy as np
from NN_Classifier import NN
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt(open('NN_Data.txt',"rt"),delimiter=" ")
np.random.shuffle(data)       #Randomnly shuffles the data
```

```python
#Creates instance of class NN
train=NN()
```

```python
#Separates data into X and Y and normalizes X
X=data[:,:2]
Y=data[:,-1:]
X=train.normalize(X)
```

```python
#dividing into training data and test data in the ratio 3:1
test_len=X.shape[0]//4
X_test=X[:test_len,:]
X_train=X[test_len:,:]

Y_test=Y[:test_len]
Y_train=Y[test_len:]
```

```python
# VISUALIZATION of training data
plt.scatter(X_train[:,0],X_train[:,1],c=Y_train.flatten())
plt.xlabel("training dataset 1")
plt.ylabel("training dataset 2")
plt.title("Training Data")
plt.show()
```

```python
#Initializing parameters
size = np.array([25])  #hidden_layer_sizes
Layers = 2   # No of hidden layers
```

```python
#fits the training set and internally optimizes the values of weights and bias 
train.fit(X_train,Y_train,size,Layers)
```

```python

#predicts the values of the test data
Y_pred = train.pred(X_test)
```

```python
# VISUALIZATION of predicted data
plt.scatter(X_test[:,0],X_test[:,1],c=Y_pred)
plt.xlabel("Testing dataset 1")
plt.ylabel("Testing dataset 2")
plt.title("Testing Data")
plt.show()
```

```python
#calculating accuracy of my model
Acc = train.accuracy(Y_test,Y_pred)
print("Accuracy of my model: ",Acc,"%")
```

```python
#calculating accuracy of sklearn model
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs',batch_size=25,hidden_layer_sizes=(128))
clf.fit(X_train, np.hstack(Y_train))
y_pred=clf.predict(X_test)
A=(y_pred==Y_test.flatten())
acc=np.mean(A)*100
print("accuracy of sklearn: ",acc,"%")
```
